```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
#define state
# the state we define is actually a dictionary in which we can tell our dtype
class BMIState(TypedDict):
    height_m: float
    weight_kg: float
    bmi: float
    category: str
```

```python
 def calculateBMI(state : BMIState) ->BMIState:
  weight_kg = state['weight_kg']
  height_m = state['height_m']
  bmi = weight_kg/(height_m*height_m)
  state['bmi'] = round(bmi,3)
  return state
```

```python
def label_BMI(state: BMIState) -> BMIState:
    bmi = state['bmi']
    if bmi < 18.5:
        state['category'] = 'Underweight'
    elif 18.5 <= bmi < 25:
        state['category'] = 'Normal'
    elif 25 <= bmi < 30:
        state['category'] = 'Overweight'
    else:
        state['category'] = 'Obese'
    return state
```

```python
# defining ouy graph

graph = StateGraph(BMIState)

# adding nodes to graph
graph.add_node("calculated BMI",calculateBMI)
graph.add_node("label_BMI",label_BMI)

# adding edges to our graph
graph.add_edge(START, "calculated BMI")
graph.add_edge("calculated BMI" , "label_BMI")
graph.add_edge( "label_BMI", END)
# compiling graph

# execution
workflow = graph.compile()
```

```python
initial = {'weight_kg': 70, 'height_m': 1.75}

final = workflow.invoke(initial)

print(final)
```
